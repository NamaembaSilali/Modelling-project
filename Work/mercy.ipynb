### DATA ANALYSIS
This section focuses on understanding the data by exploring the data visually to identify patterns, relationships and potential issues.

### Histograms
Histograms are a great way to visualize the distribution of data. Below are histograms for price, bedrooms, sqft_living, and yr_built to understand the distribution of each variable.

```python
# Histograms for numerical columns
plt.figure(figsize=(12, 6))

# Histogram for price
plt.subplot(2, 2, 1)
sns.histplot(data['price'], kde=True)
plt.title('Price Distribution')

# Histogram for sqft_living
plt.subplot(2, 2, 2)
sns.histplot(data['sqft_living'], kde=True)
plt.title('Sqft Living Distribution')

# Histogram for yr_built
plt.subplot(2, 2, 3)
sns.histplot(data['yr_built'], kde=True)
plt.title('Year Built Distribution')

plt.tight_layout()
plt.show()
```

### BOX PLOTS
Box plots are used to show the distribution of a dataset. They are useful for comparing datasets. Box plots are made up of a box, a median line, and whiskers. The box represents the interquartile range (IQR), which is the range of values between the first and third quartiles. The median line is the middle of the dataset. The whiskers extend from the box to the minimum and maximum values in the dataset. The whiskers can be extended to the minimum and maximum values, or they can be extended to the minimum and maximum values that are not outliers. Outliers are values that are more than 1.5 times the IQR above the third quartile or more than 1.5 times the IQR below the first quartile. 
Box plots for bedrooms, condition, and yr_built to identify outliers and understand the distribution of these variables.

```python
# Box plots for numerical columns
plt.figure(figsize=(12, 6))

# Box plot for price
plt.subplot(2, 2, 1)
sns.boxplot(data['price'])
plt.title('Price Box Plot')

# Box plot for sqft_living
plt.subplot(2, 2, 2)
sns.boxplot(data['sqft_living'])
plt.title('Sqft Living Box Plot')

# Box plot for yr_built
plt.subplot(2, 2, 3)
sns.boxplot(data['yr_built'])
plt.title('Year Built Box Plot')

# Box plot for bedrooms
plt.subplot(2, 2, 4)
sns.boxplot(data['bedrooms'])
plt.title('Bedrooms Box Plot')

plt.tight_layout()
plt.show()
```

Count plots for bedrooms and condition to visualize the frequency of different categories.

```python
# Count plot for 'bedrooms'
plt.figure(figsize=(8, 6))
sns.countplot(data=data_final, x='bedrooms')
plt.title('Count of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()

# Count plot for 'condition'
plt.figure(figsize=(8, 6))
sns.countplot(data=data_final, x='condition')
plt.title('Count of Condition')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.show()
```

### Scatter plots
Price vs Sqft Living: This plot typically shows a strong positive correlation. As the square footage of a house increases, the price tends to increase as well. However, there might be outliers or variations due to factors like location, amenities, or condition.
Price vs Bedrooms: The relationship might be positive, but it's less pronounced than with sqft_living. More bedrooms generally correlate with higher prices, but other factors like house size and location play a more significant role.
Price vs Yr Built: This plot might show a trend, but it's complex. Newer houses often command higher prices, but factors like architectural style, location, and renovations can influence the price.
### Box Plots
Price by Condition: This plot helps understand how the price varies across different house conditions. Generally, houses in better condition tend to have higher median prices and smaller ranges. However, outliers in higher-condition categories might indicate luxury homes or unique properties.
### Correlation Matrix
The correlation matrix quantifies the linear relationships between numerical variables. Values close to 1 indicate a strong positive correlation, values close to -1 indicate a strong negative correlation, and values close to 0 indicate a weak or no correlation.

```python
# Scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(data=data_final, x='sqft_living', y='price')
plt.title('Price vs Sqft Living')

plt.subplot(1, 3, 2)
sns.scatterplot(data=data_final, x='bedrooms', y='price')
plt.title('Price vs Bedrooms')

plt.subplot(1, 3, 3)
sns.scatterplot(data=data_final, x='yr_built', y='price')
plt.title('Price vs Yr Built')

plt.tight_layout() 
plt.show()

# Box plots of price grouped by condition
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_final, x='condition', y='price')
plt.title('Price by Condition')
plt.show()

# Correlation matrix
correlation_matrix = data_final.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
